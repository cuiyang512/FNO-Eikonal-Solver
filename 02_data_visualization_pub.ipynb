## Import library

```python
import matplotlib.pyplot as plt 
import numpy as np 
import os
import random 

random.seed(202510)
```

## Load the training and testing data in different velocity families

```python
train_data = np.load('./data/train_vel.npy')
test_data = np.load('./data/test_vel.npy')

print('train_data shape:', train_data.shape, 'test_data shape:', test_data.shape)
```

## Training Data

```python
num_profiles = 10

for ii in range (0, 3):
    
    random_indices = random.sample(range(train_data.shape[0]), num_profiles)
    
    # Create a figure with 2x5 subplots
    fig, axes = plt.subplots(2, 5, figsize=(12, 6))
    axes = axes.flatten()
    
    for i, idx in enumerate(random_indices):
        profile = train_data[idx, ii, :, :]
        
        im = axes[i].imshow(profile, cmap='jet', aspect='equal')
        axes[i].set_title(f'Profile {idx}')
        axes[i].set_xticks([])
        axes[i].set_yticks([])
    
    plt.tight_layout()
    plt.show()
```

## Testing Data

```python
for ii in range (0, 1):
    # Select 10 random indices
    random_indices = random.sample(range(test_data.shape[0]), num_profiles)
    
    # Create a figure with 2x5 subplots
    fig, axes = plt.subplots(2, 5, figsize=(12, 6))
    axes = axes.flatten()
    
    for i, idx in enumerate(random_indices):
        profile = test_data[idx, ii, :, :]
        
        im = axes[i].imshow(profile, cmap='jet', aspect='equal')
        axes[i].set_title(f'Profile {idx}')
        axes[i].set_xticks([])
        axes[i].set_yticks([])
    
    plt.tight_layout()
    plt.show()
```
